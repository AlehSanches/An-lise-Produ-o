```python
import pandas as pd
import matplotlib.pyplot as plt

# Ler o arquivo CSV
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/peixes.csv', sep=';')

# Agrupar a produção por tipo de peixe e mês
producao_mensal = df.groupby(['PEIXE', 'MES'])['PRODUCAO'].sum().unstack()

# Criar o gráfico de barras
producao_mensal.plot(kind='bar', figsize=(10, 6))

# Adicionar título e rótulos
plt.title('Produção Mensal por Tipo de Peixe')
plt.xlabel('Tipo de Peixe')
plt.ylabel('Produção (kg)')

# Mostrar o gráfico
plt.xticks(rotation=45) # Rotacionar rótulos do eixo x para melhor visualização
plt.tight_layout() # Ajustar layout para evitar cortes
plt.show()
```
